```python
# do a fourrier transform of an image and take it back to the original image

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import fftpack

# read the image
img = mpimg.imread(R'C:\Users\maxim\Desktop\before comp 14.6m 2.png')

# do the fourrier transform
img_fft = fftpack.fft2(img)

# do the inverse fourrier transform
img_new = fftpack.ifft2(img_fft)

# plot the original image
plt.figure()
plt.imshow(img, cmap='gray')
plt.title('Original image')

# plot the fourrier transform
plt.figure()
plt.imshow(np.abs(img_fft), cmap='gray')
plt.title('Fourrier transform')

# plot the new image
plt.figure()
plt.imshow(np.abs(img_new), cmap='gray')
plt.title('New image')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import fftpack

# read the image
img = mpimg.imread(R'C:\Users\maxim\Desktop\before comp 14.6m 2.png')

# do the fourrier transform
img_fft = fftpack.fft2(img)
```

```python
# can you plot other thing like the phase of the fourrier transform?

img_fft_phase = np.angle(img_fft)

plt.figure()
plt.imshow(img_fft_phase, cmap='gray')
plt.title('Fourrier transform phase')

```

```python
# can you plot other thing like frequency of the fourrier transform?

img_fft_freq = np.abs(img_fft)

plt.figure()
plt.imshow(img_fft_freq, cmap='gray')
plt.title('Fourrier transform frequency')

# can you plot other thing like the real part of the fourrier transform?

img_fft_real = np.real(img_fft)

plt.figure()
plt.imshow(img_fft_real, cmap='gray')
plt.title('Fourrier transform real part')

# can you plot other thing like the imaginary part of the fourrier transform?

img_fft_imag = np.imag(img_fft)

plt.figure()
plt.imshow(img_fft_imag, cmap='gray')
plt.title('Fourrier transform imaginary part')

```

```python
#plot frequency graph

plt.figure()
plt.plot(img_fft_freq[0,:])
plt.title('Fourrier transform frequency graph')

#plot phase graph

plt.figure()
plt.plot(img_fft_phase[0,:])
plt.title('Fourrier transform phase graph')


```

```python
# display the graph with the fourrier transform

plt.figure()
plt.imshow(np.log(np.abs(img_fft)), cmap='gray')
plt.title('Fourrier transform with log')

# display the graph with the fourrier transform frequency and the phase

plt.figure()
plt.imshow(np.log(np.abs(img_fft_freq)), cmap='gray')
plt.title('Fourrier transform with log and phase')


freqCentered = np.log(np.abs(np.fft.fftshift(img_fft_freq)))
plt.figure()
plt.imshow(freqCentered, cmap='gray')
plt.title('Fourrier 1')

# return to normal state by removing the np.log and np.abs

FreqReturned= np.fft.ifftshift(np.fft.fftshift(img_fft_freq))


plt.figure()
plt.imshow(np.log(np.abs(np.fft.ifftshift(np.fft.fftshift(img_fft_freq)))), cmap='gray')
plt.title('Fourrier 1')


# display the graph with the fourrier transform phase 

plt.figure()
plt.imshow(np.log(np.abs(img_fft_phase)), cmap='gray')
plt.title('Fourrier transform with log and phase')

```
